```python
import numpy as np
from PyAstronomy import pyasl
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#Latex font
import matplotlib as mpl
from matplotlib import rc
plt.rc('text', usetex=True)
plt.rc('font', family='serif',size=16)
```

```python
orbit = pyasl.KeplerEllipse(a=8.19, per=10220 , e=0.39, tau = 2453020, Omega=0.0, i=10, w=-78)
t = np.linspace(0, 10220, 500)
#print(t)
pos_a = orbit.xyzPos(t)
#print(pos)
```

```python
fig = plt.subplots(1, 1, sharex=True, sharey=False, figsize=(15, 5), tight_layout=True)

plt.plot(0, 0, 'bo', markersize=9, label="DP Leo")
plt.plot(pos_a[0, 1], pos_a[0, 0], 'r*',markersize=10,label="Periapsis")
plt.plot(pos_a[::, 1], pos_a[::, 0], 'r-',label='K = 711 sec')

# Put a legend to the right of the current axis
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel('X (AU)')
plt.ylabel('Y (AU)')
#plt.legend(loc="best")
#plt.title('Orbital Simulation')
plt.show()
```
